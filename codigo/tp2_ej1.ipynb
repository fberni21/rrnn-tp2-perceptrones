# Ejercicio 1

Implemente un perceptrón simple que aprenda la función lógica AND y la función lógica OR, de 2 y de 4 entradas. Muestre la evolución del error durante el entrenamiento. Para el caso de 2 dimensiones, grafique la recta discriminadora y todos los vectores de entrada de la red.

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(4574)
```

```python
def sign(X):
    return np.where(X >= 0, 1, -1)
```

```python
class Perceptron:
    def __init__(self):
        self.weights = None
        self.errors = None
        self.num_inputs = None
        
    def predict(self, X):
        X = np.atleast_2d(X)

        if X.shape[1] == self.num_inputs - 1:
            X = np.c_[X, np.ones(X.shape[0])] 

        return self._predict(X)

    def fit(self, X, y, epochs=10, learning_rate=1e-3, tol=0.1):
        eta = learning_rate
        X = np.c_[X, np.ones(X.shape[0])]
        Np, N = X.shape
        self.num_inputs = N
        self.weights = np.random.uniform(-1, 1, size=N)
        self.errors = []

        for _ in range(epochs):
            for i in np.random.permutation(Np):
                xi = X[i, :]
                yi = y[i]
                yi_pred = self._predict(xi)
                delta_w = eta * xi * (yi - yi_pred)
                self.weights += delta_w
                error = self.error(X, y)
                self.errors.append(error)
                if error < tol:
                    return

    def error(self, X, y):
        y_pred = self.predict(X)
        return np.mean((y - y_pred)**2)

    def _predict(self, X):
        return sign(np.dot(self.weights, X.T))
```

```python
X2 = np.array([
    [-1, -1],
    [-1,  1],
    [ 1, -1],
    [ 1,  1]
])
y2_and = np.array([-1, -1, -1, 1])
y2_or  = np.array([-1,  1,  1, 1])
```

```python
def plot_mse(model, filename=None):
    plt.figure(figsize=(6,4))
    plt.plot(model.errors, marker='.')
    plt.xlabel('Iteración')
    plt.ylabel('Error cuadrático medio')
    plt.grid()
    plt.tight_layout()
    if filename:
        plt.savefig(filename)
    plt.show()
```

```python
def plot_decision_boundary_2d(model, X, filename=None):
    plt.figure(figsize=(6,6))
    
    y_pred = model.predict(X)
    plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='bwr', s=200, edgecolors='k')
    
    m = -model.weights[1]/model.weights[0]
    b = -model.weights[2]/model.weights[0]
    x = np.linspace(-1.5, 1.5, 2)
    y = m*x + b
    plt.plot(x, y, '--g', label='Frontera de decisión')
    
    plt.xlabel('Entrada $X_1$')
    plt.ylabel('Entrada $X_2$')
    
    plt.xlim([-1.5, 1.5])
    plt.ylim([-1.5, 1.5])
    plt.legend()
    plt.grid()
    plt.tight_layout()
    if filename:
        plt.savefig(filename)
    plt.show()
```

```python
model = Perceptron()
```

```python
model.fit(X2, y2_and, learning_rate=1e-2, epochs=20)
plot_mse(model, filename="and2d_mse.pdf")
```

```python
plot_decision_boundary_2d(model, X2, filename="and_boundary.pdf")
```

```python
model.fit(X2, y2_or, learning_rate=1e-2, epochs=20)
plot_mse(model, filename="or2d_mse.pdf")
```

```python
plot_decision_boundary_2d(model, X2, filename="or_boundary.pdf")
```

```python
X4 = np.array(list(itertools.product([-1, 1], repeat=4)))
y4_and = np.where(np.all(X4 == 1, axis=1), 1, -1)
y4_or  = np.where(np.any(X4 == 1, axis=1), 1, -1)
```

```python
model.fit(X4, y4_and, learning_rate=1e-2, epochs=20)
plot_mse(model, filename="and4d_mse.pdf")
```

```python
model.fit(X4, y4_or, learning_rate=1e-2, epochs=20)
plot_mse(model, filename="or4d_mse.pdf")
```
