# Ejercicio 1

Implemente un perceptrón simple que aprenda la función lógica AND y la función lógica OR, de 2 y de 4 entradas. Muestre la evolución del error durante el entrenamiento. Para el caso de 2 dimensiones, grafique la recta discriminadora y todos los vectores de entrada de la red.

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(4574)
```

```python
def sign(X):
    return np.where(X >= 0, 1, -1)
```

```python
class Perceptron:
    def __init__(self):
        self.weights = None
        self.errors = None
        
    def predict(self, X):
        if X.ndim == 1:
            X = X.reshape(1, -1)
            
        if X.shape[1] == self.weights.shape[0] - 1:
            X= np.c_[X, np.ones(X.shape[0])] 
            
        return sign(np.dot(self.weights, X.T))

    def fit(self, X, y, epochs=10, learning_rate=1e-3, tol=0.1):
        eta = learning_rate
        Np, N = X.shape
        X_ext = np.c_[X, np.ones(Np)]
        self.weights = np.random.rand(N + 1) * 2 - 1
        self.errors = []

        for _ in range(epochs):
            for i in np.random.permutation(Np):
                xi = X_ext[i, :]
                yi = y[i]
                yi_pred = sign(np.dot(self.weights, xi))
                delta_w = eta * xi * (yi - yi_pred)
                self.weights += delta_w
                error = self.error(X, y)
                self.errors.append(error)
                if error < tol:
                    return

    def error(self, X, y):
        y_pred = self.predict(X)
        return np.mean((y - y_pred)**2)
```

```python
X_and = np.array([
    [-1, -1],
    [-1,  1],
    [ 1, -1],
    [ 1,  1]
])
y_and = np.array([-1, -1, -1, 1])
```

```python
model = Perceptron()
model.fit(X_and, y_and, learning_rate=1e-2, epochs=20)
```

```python
plt.figure()
plt.plot(model.errors, marker='x')
plt.xlabel('Iteración')
plt.ylabel('Error cuadrático medio')
plt.grid()
plt.show()
```

# 

```python
plt.figure(figsize=(6,6))

y_pred = model.predict(X_and)
plt.scatter(X_and[:,0], X_and[:,1], c=y_pred, cmap='bwr', s=200, edgecolors='k')

m = -model.weights[1]/model.weights[0]
b = -model.weights[2]/model.weights[0]
x = np.linspace(-1.5, 1.5, 2)
y = m*x + b
plt.plot(x, y, '-g', label='Frontera de decisión')

plt.xlabel('Entrada $X_1$')
plt.ylabel('Entrada $X_2$')

plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.legend()
plt.grid()
plt.show()
```
