# Ejercicio 2

Determine numéricamente cómo varía la capacidad del perceptrón simple en función del número de patrones enseñados.

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(110007)
```

```python
def sign(X):
    return np.where(X >= 0, 1, -1)
```

```python
class Perceptron:
    def __init__(self):
        self.weights = None
        self.errors = None
        self.num_inputs = None
        
    def predict(self, X):
        X = np.atleast_2d(X)

        if X.shape[1] == self.num_inputs - 1:
            X = np.c_[X, np.ones(X.shape[0])] 

        return self._predict(X)

    def fit(self, X, y, epochs=10, learning_rate=0.1, tol=0.1):
        eta = learning_rate
        X = np.c_[X, np.ones(X.shape[0])]
        Np, N = X.shape
        self.num_inputs = N
        self.weights = np.random.uniform(-1, 1, size=N)
        self.errors = [self.error(X, y)]

        for _ in range(epochs):
            for i in np.random.permutation(Np):
                xi = X[i, :]
                yi = y[i]
                yi_pred = self._predict(xi)
                delta_w = eta * xi * (yi - yi_pred)
                self.weights += delta_w
                error = self.error(X, y)
                self.errors.append(error)
                if error < tol:
                    return True
        return False

    def error(self, X, y):
        y_pred = self.predict(X)
        return np.mean((y - y_pred)**2)

    def _predict(self, X):
        return sign(np.dot(self.weights, X.T))
```

```python
N = 30
Np_max = 3*N
trials = 20

model = Perceptron()
probs = np.zeros(Np_max)
for Np in range(1, Np_max+1):
    print(f'Np={Np:02d}', end='\r')
    for trial in range(trials):
        X = np.random.uniform(-1, 1, size=(Np, N))
        y = np.random.choice([-1, 1], size=Np)
        converged = model.fit(X, y, learning_rate=0.01, epochs=1000)
        if converged:
            probs[Np-1] += 1

probs /= trials
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.arange(1, Np_max+1)/N, probs, marker='.', label='$N = 30$')
plt.vlines(2, -0.03, 1.03, 'tab:red', linestyles='dashed', label='Límite teórico ($N\\to\\infty$)')
plt.xlabel('$N_p/N$')
plt.ylabel('Probabilidad de convergencia')
plt.ylim([-0.05, 1.05])
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig('../informe/img/ej2/capacidad.svg')
plt.show()
```
